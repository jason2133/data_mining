```python
# https://www.w3schools.com/python/python_ml_hierarchial_clustering.asp
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.cluster.hierarchy.linkage.html

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
```

```python
# /content/drive/MyDrive/Colab Notebooks/diabetes.xls
# /content/drive/MyDrive/diabetes_data.xlsx

data = pd.read_excel('/content/drive/MyDrive/diabetes_data.xlsx')
data
```

```python
data = data.drop(['ID'], axis=1)
data
```

```python
# Ward

plt.figure(figsize=(30, 30))
linkage_data_ward = linkage(data, method='ward', metric='euclidean')
dendrogram(linkage_data_ward)
plt.show()
```

```python
plt.figure(figsize=(10, 10))
dendrogram(linkage_data_ward)
plt.show()
```

```python
print(linkage_data_ward)
```

```python
# Complete

plt.figure(figsize=(30, 30))
linkage_data_complete = linkage(data, method='complete', metric='euclidean')
dendrogram(linkage_data_complete)
plt.show()
```

```python
plt.figure(figsize=(10, 10))
dendrogram(linkage_data_complete)
plt.show()
```

```python
# Single

plt.figure(figsize=(30, 30))
linkage_data_single = linkage(data, method='single', metric='euclidean')
dendrogram(linkage_data_single)
plt.show()
```

```python
plt.figure(figsize=(10, 10))
dendrogram(linkage_data_single)
plt.show()
```

```python
# Average

plt.figure(figsize=(30, 30))
linkage_data_average = linkage(data, method='average', metric='euclidean')
dendrogram(linkage_data_average)
plt.show()
```

```python
plt.figure(figsize=(10, 10))
dendrogram(linkage_data_average)
plt.show()
```

```python
# Centroid

plt.figure(figsize=(30, 30))
linkage_data_centroid = linkage(data, method='centroid', metric='euclidean')
dendrogram(linkage_data_centroid)
plt.show()
```

```python
plt.figure(figsize=(10, 10))
dendrogram(linkage_data_centroid)
plt.show()
```

```python
# Median

plt.figure(figsize=(30, 30))
linkage_data_median = linkage(data, method='median', metric='euclidean')
dendrogram(linkage_data_median)
plt.show()
```

```python
plt.figure(figsize=(10, 10))
dendrogram(linkage_data_median)
plt.show()
```
