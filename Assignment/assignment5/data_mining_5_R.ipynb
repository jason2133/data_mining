### Hierarchical Clustering
- https://www.rdocumentation.org/packages/stats/versions/3.6.2/topics/hclust

```r
cluster_complete = hclust(dist(data_scale), method='complete')
plot(cluster_complete)
```

```r
cluster_average = hclust(dist(data_scale), method='average')
plot(cluster_average)
```

```r
cluster_single = hclust(dist(data_scale), method='single')
plot(cluster_single)
```

```r
cluster_wardd = hclust(dist(data_scale), method='ward.D')
plot(cluster_wardd)
```

```r
cluster_wardd2 = hclust(dist(data_scale), method='ward.D2')
plot(cluster_wardd2)
```

```r
cluster_mcquitty = hclust(dist(data_scale), method='mcquitty')
plot(cluster_mcquitty)
```

```r
cluster_median = hclust(dist(data_scale), method='median')
plot(cluster_median)
```

```r
cluster_centroid = hclust(dist(data_scale), method='centroid')
plot(cluster_centroid)
```

### Cut the dendrogram

```r
cluster_complete.cut = cutree(cluster_complete, k=3)
cluster_complete.cut
```

```r
cluster_complete.cut2 = cutree(cluster_complete, k=7)
cluster_complete.cut2
```

### Showing the result

```r
h.seg1 = data_scale[cluster_complete.cut==1,]
h.seg2 = data_scale[cluster_complete.cut==2,]
h.seg3 = data_scale[cluster_complete.cut==3,]
```

```r
pie(table(cluster_complete.cut),main="number of observations in segment")
```

```r
h.mean = rbind(apply(h.seg1,2,mean),apply(h.seg2,2,mean),apply(h.seg3,2,mean))
```

```r
rownames(h.mean) = c(1,2,3)
```

```r
h.mean
```

```r
dist(h.mean, method = "euclidean", diag = TRUE)
```
