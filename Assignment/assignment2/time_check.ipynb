```python
# Ctrl + Alt + Enter

import numpy as np

# Month
month_list = [i for i in range(1, 13)]

# Day
day_list = [i for i in range(1, 31)]

# Hour
hour_list = [i for i in range(1, 25)]
```

```python
def month_x(x):
    ans = np.sin((360 / 12) * x / 180 * np.pi)
    return ans

def month_y(x):
    ans = np.cos((360 / 12) * x / 180 * np.pi)
    return ans
```

```python
def day_x(x):
    ans = np.sin((360 / 30) * x / 180 * np.pi)
    return ans

def day_y(x):
    ans = np.cos((360 / 30) * x / 180 * np.pi)
    return ans
```

```python
def hour_x(x):
    ans = np.sin((360 / 24) * x / 180 * np.pi)
    return ans

def hour_y(x):
    ans = np.cos((360 / 24) * x / 180 * np.pi)
    return ans
```

```python
# Month
month_list = [i for i in range(1, 13)]

month_x_list = []
month_y_list = []

for i in range(len(month_list)):
    month_x_list.append(month_x(month_list[i]))
    month_y_list.append(month_y(month_list[i]))

print('Month X')
print(month_x_list)
print('-' * 50)
print('Month Y')
print(month_y_list)
```

```python
# day
day_list = [i for i in range(1, 31)]

day_x_list = []
day_y_list = []

for i in range(len(day_list)):
    day_x_list.append(day_x(day_list[i]))
    day_y_list.append(day_y(day_list[i]))

print('day X')
print(day_x_list)
print('-' * 50)
print('day Y')
print(day_y_list)
```

```python
# Hour
hour_list = [i for i in range(1, 25)]

hour_x_list = []
hour_y_list = []

for i in range(len(hour_list)):
    hour_x_list.append(hour_x(hour_list[i]))
    hour_y_list.append(hour_y(hour_list[i]))

print('hour X')
print(hour_x_list)
print('-' * 50)
print('hour Y')
print(hour_y_list)
```
